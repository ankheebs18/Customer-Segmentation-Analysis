```python
# Average Order Value (AOV) Trend Over Time
aov_over_time = df.groupby('YearMonth')['TotalPrice'].mean()
plt.figure(figsize=(12, 6))
aov_over_time.plot(kind='line', marker='o', color='purple')
plt.title('Average Order Value (AOV) Trend Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Average Order Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# RFM Analysis for customer segmentation
# Recency: Number of days since the last purchase
# Frequency: Total number of purchases
# Monetary: Total money spent
```

```python
# Calculate Recency
current_date = df['InvoiceDate'].max()
df['Recency'] = (current_date - df['InvoiceDate']).dt.days
recency_df = df.groupby('CustomerID')['Recency'].min().reset_index()
```

```python
# Calculate Frequency
frequency_df = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
frequency_df.columns = ['CustomerID', 'Frequency']
```

```python
# Calculate Monetary
monetary_df = df.groupby('CustomerID')['TotalPrice'].sum().reset_index()
monetary_df.columns = ['CustomerID', 'Monetary']
```

```python
# Combine RFM metrics
rfm_df = recency_df.merge(frequency_df, on='CustomerID').merge(monetary_df, on='CustomerID')
```

```python
# Function to calculate RFM score with dynamic binning
def calculate_rfm_score(df, column, labels):
    try:
        return pd.qcut(df[column], 4, labels=labels, duplicates='drop')
    except ValueError:
        unique_values = df[column].nunique()
        if unique_values < 2:  # Edge case handling for columns with 1 unique value
            return pd.Series([labels[0]] * len(df), index=df.index)
        else:
            # Adjust labels to be one fewer than bins
            adjusted_labels = labels[:unique_values - 1]
            bins = unique_values if unique_values < len(labels) else len(labels)
            return pd.qcut(df[column], bins, labels=adjusted_labels, duplicates='drop')
```

```python
# Ensure labels match the number of bins dynamically
unique_recency = rfm_df['Recency'].nunique()
unique_frequency = rfm_df['Frequency'].nunique()
unique_monetary = rfm_df['Monetary'].nunique()
```

```python
# RFM Score
rfm_df['R_Score'] = calculate_rfm_score(rfm_df, 'Recency', ['1', '2', '3'][:unique_recency])
rfm_df['F_Score'] = calculate_rfm_score(rfm_df, 'Frequency', ['3', '2', '1'][:unique_frequency])
rfm_df['M_Score'] = calculate_rfm_score(rfm_df, 'Monetary', ['3', '2', '1'][:unique_monetary])
```

```python
# Combine RFM scores
rfm_df['RFM_Score'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)
```

```python
# Display the RFM scores
print(rfm_df.head())
```

```python
# Customer Segmentation
# We can use clustering techniques like K-means to segment the customers
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Scale the RFM values
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_df[['Recency', 'Frequency', 'Monetary']])
```

```python
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
# Check the current dtype of 'Cluster'
print(rfm_df['Cluster'].dtype)
```

```python
# Convert 'Cluster' to categorical if it's not already
if not pd.api.types.is_categorical_dtype(rfm_df['Cluster']):
    rfm_df['Cluster'] = rfm_df['Cluster'].astype('category')
```

```python
# Now you can access .cat.codes safely
rfm_df['Cluster_numeric'] = rfm_df['Cluster'].cat.codes
```

```python
# Plot the clusters
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Recency', y='Monetary', hue='Cluster', data=rfm_df, palette='viridis')
plt.title('Customer Segments based on RFM Analysis')
plt.xlabel('Recency')
plt.ylabel('Monetary')
plt.show()
```

```python
# Save the RFM and cluster information to a CSV file
rfm_df.to_csv('customer_segments.csv', index=False)
```
